## Introdução à biblioteca Matplotlib

O Matplotlib é uma biblioteca de plotagem Python 2D que produz números de qualidade de publicação em uma variedade de formatos impressos e ambientes interativos em plataformas. Ele tenta facilitar coisas simples e difíceis. Você pode gerar gráficos, histogramas, espectros de poder, gráficos de barras, gráficos de erros, diagramas de dispersão, etc., com apenas algumas linhas de código. 

Este é apenas um tutorial introdutório a esta ferramenta tão poderosa e bastante utilizada na ciência de dados.


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

À seguir, mostraremos plotagens de algumas funções conhecidas:

  #### Plotagem da função:
  $$ f(x) = \sqrt{x} $$

```python
x = [x for x in range(20)]
y = [y**2 for y in range(20)]
plt.plot(x,y, marker = 'o');
```

####  Plotagem da função:
  $$ f(x) = 2^{x} $$

Adicionamos uma propriedade de mostrar o gráfico em formato de grade.

```python
u = [u for u in range(10)]
v = [ 2**v for v in range(10)]
plt.plot(u,v, marker = 'o');
plt.grid()
```

####  Plotagem da função:
  $$ f(x) = x $$

```python
z = [z for z in range(20)]
w = [w for w in range(20)]
plt.scatter(z,w, marker='x')
```

```python
anos = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
pib = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
plt.plot(anos, pib, marker='o')
plt.title('PIB')
plt.xlabel('Ano')
plt.ylabel(u'Bilhões de R$')
plt.grid()
```

#### Gráfico de Barras

```python
x = [2,4,6,8,10]
y = [6,7,8,2,4]
plt.bar(x, y, label = 'Barras', color = 'r')
plt.legend()
plt.show()
```

```python
x2 = [1,3,5,7,9]
y2 = [7,8,2,4,2]
plt.bar(x, y, label = 'Barras1', color = 'r')
plt.bar(x2, y2, label = 'Barras2', color = 'y')
plt.legend()
plt.show()
```

#### Scatterplot

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,6,8,4,8]
plt.scatter(x, y, label = 'Pontos', color = 'k', marker = 'o')
plt.legend()
plt.show()
```

#### Pie Chart

```python
fatias = [7, 2, 2, 13]
atividades = ['dormir','comer','trabalhar','passear']
plt.pie(fatias, labels = atividades)
plt.show()
```
